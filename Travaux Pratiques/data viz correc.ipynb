```python
import matplotlib.pyplot as plt
import random
plt.style.use("ggplot")
```

```python

# Exercice 1 : Tracer un graphique linéaire
plt.figure(figsize=(15, 10))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [random.randint(1000, 5000) for _ in range(12)]

plt.plot(months, sales)
plt.xlabel('Mois')
plt.ylabel('Ventes')
plt.title('Ventes mensuelles')
plt.show()
```

```python
# Exercice 2 : Créer un histogramme
plt.figure(figsize=(15, 10))
scores = [random.randint(50, 100) for _ in range(100)]
plt.hist(scores, bins=10, edgecolor='k')
plt.xlabel('Scores')
plt.ylabel('Nombre d\'étudiants')
plt.title('Distribution des scores')
plt.show()
```

```python
# Exercice 3 : Tracer un diagramme en barres
plt.figure(figsize=(15, 10))
brands = ['Toyota', 'Honda', 'Ford', 'Chevrolet', 'Nissan']
car_count = [random.randint(100, 500) for _ in range(5)]

plt.bar(brands, car_count)
plt.xlabel('Marque de voiture')
plt.ylabel('Nombre de voitures')
plt.title('Nombre de voitures par marque')
plt.show()
```

```python
# Exercice 4 : Créer un nuage de points
plt.figure(figsize=(15, 10))
x = [random.randint(1, 50) for _ in range(50)]
y = [random.randint(1, 50) for _ in range(50)]

plt.scatter(x, y)
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Nuage de points')
plt.show()
```

```python
# Exercice 5 : Construire un diagramme circulaire
plt.figure(figsize=(10, 7))
products = ['Produit A', 'Produit B', 'Produit C', 'Produit D']
sales = [random.randint(10, 100) for _ in range(4)]

plt.pie(sales, labels=products, autopct='%1.1f%%')
plt.title('Répartition des ventes par produit')
plt.show()
```

```python
# Exercice 6 : Créer un graphique à barres empilées
plt.figure(figsize=(15, 10))
years = [2019, 2020, 2021]
product_a = [random.randint(100, 300) for _ in range(3)]
product_b = [random.randint(100, 300) for _ in range(3)]

plt.bar(years, product_a, label='Produit A')
plt.bar(years, product_b, label='Produit B', bottom=product_a)
plt.xlabel('Année')
plt.ylabel('Ventes')
plt.title('Ventes par année et par produit')
plt.legend()
plt.show()
```

```python
# Exercice 7 : Créer un graphique à secteurs
plt.figure(figsize=(15, 10))
categories = ['Logement', 'Alimentation', 'Transport', 'Loisirs', 'Autres']
expenses = [random.randint(500, 3000) for _ in range(5)]

plt.pie(expenses, labels=categories, autopct='%1.1f%%')
plt.title('Répartition des dépenses mensuelles')
plt.show()
```

```python
# Exercice 8 : Tracer un graphique de lignes avec plusieurs lignes
plt.figure(figsize=(15, 10))
years = [2010, 2011, 2012, 2013, 2014]
product_a = [random.randint(100, 300) for _ in range(5)]
product_b = [random.randint(100, 300) for _ in range(5)]

plt.plot(years, product_a, label='Produit A')
plt.plot(years, product_b, label='Produit B')
plt.xlabel('Année')
plt.ylabel('Ventes')
plt.title('Ventes par année et par produit')
plt.legend()
plt.show()
```

```python
# Exercice 9 : Créer un graphique en boîte
plt.figure(figsize=(15, 10))
data = [random.randint(50, 100) for _ in range(100)]
plt.boxplot(data)
plt.title('Boîte à moustaches des scores')
plt.show()
```
