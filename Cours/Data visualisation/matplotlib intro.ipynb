<h3>C'est quoi la visualisation de données?</h3>

La visualisation de données est le processus de représentation graphique des informations et des données afin de les rendre plus compréhensibles et significatives. Elle utilise des éléments visuels tels que des graphiques, des tableaux, des diagrammes et d'autres formes graphiques pour illustrer les tendances, les schémas et les relations entre les données.

L'objectif de la visualisation de données est de faciliter la compréhension, l'analyse et la communication des informations contenues dans les ensembles de données. Cela permet aux individus, aux entreprises et aux organisations de prendre des décisions éclairées et d'identifier des insights à partir des données.

La visualisation de données est largement utilisée dans divers domaines tels que la science des données, la business intelligence, la recherche, la finance, la médecine, et bien d'autres. Elle joue un rôle crucial dans la présentation efficace des résultats de l'analyse de données et dans la communication des conclusions aux parties prenantes concernées.

la visualisation de données présente un énorme avantages pour les analystes de données, voici quelques raisons:

**Facilité de Compréhension :** Les graphiques et les visualisations rendent les données plus accessibles et compréhensibles pour un large public, quel que soit le niveau de familiarité avec les chiffres.

**Identification de Tendances :**  Les tendances, les schémas et les relations entre les données sont souvent plus évidents lorsqu'ils sont représentés visuellement, facilitant ainsi leur identification.

**Communication Efficace :** Les images parlent souvent plus fort que les mots. Les visualisations permettent de communiquer des idées complexes de manière simple et convaincante.

**Identification d'Anomalies :**  Les valeurs aberrantes ou les anomalies sont souvent plus visibles dans une visualisation, ce qui facilite leur détection et leur gestion.

**Comparaison Facilitée :**Les visualisations permettent de comparer facilement des ensembles de données différents, mettant en évidence les similitudes et les différences.

<h4>Quelques exemples de graph</h4>

<strong>Histogramme</strong> :
    Un histogramme représente la distribution des données en regroupant les valeurs en intervalles. Il montre la fréquence de chaque intervalle. Il est très utile pour visualiser la distribution et la forme d'une variable numerique

<strong>Diagramme en barres (Bar Plot)</strong> :
 Il représente des catégories sur l'axe des x et des valeurs numériques sur l'axe des y sous forme de barres verticales ou horizontales. Idéal pour comparer des catégories discrètes

<strong>Diagramme à secteurs (Pie Chart)</strong> :
Il représente les données en fractions d'un tout sous forme de secteurs d'un cercle Utile pour montrer la composition relative de différentes catégories.

<strong>Diagramme en boîte (Box Plot)</strong> :
Il représente la distribution des données en affichant les quartiles et les valeurs aberrantes.Utile pour visualiser la dispersion et les valeurs aberrantes d'une variable numérique.

<strong>Nuage de points (Scatter Plot)</strong> :
 Il représente les valeurs de deux variables numériques sous forme de points dans un espace bidimensionnel.
Pour visualiser la relation entre deux variables continues.

<strong>Diagramme de dispersion (Pair Plot)</strong> :
Il affiche des paires de relations entre plusieurs variables numériques dans une matrice de graphiques de dispersion.Utile pour identifier les tendances et les relations dans un ensemble de données.

<strong>Diagramme en violon (Violin Plot)</strong> :
 Il combine un diagramme en boîte avec une estimation de la densité de probabilité de la distribution des données.Utiliser pour visualiser la distribution et la densité de probabilité d'une variable numérique.

<strong>Graphique de densité (Density Plot)</strong> :
Il représente la distribution de probabilité d'une variable continue en utilisant des courbes de densité.
utiliser visualiser la distribution de probabilité d'une variable continue.



**Introduction à Matplotlib**

Matplotlib est une bibliothèque de visualisation de données en Python qui offre un large éventail d'outils pour créer des graphiques et des visualisations de données de manière efficace et personnalisée. Que ce soit des graphiques simples ou des visualisations complexes, Matplotlib est largement utilisé dans la communauté des data scientists et des analystes de données.

**Pourquoi Utiliser Matplotlib ?**

1. **Facilité d'Utilisation :** Matplotlib est relativement facile à prendre en main pour créer des graphiques de base. Ses fonctionnalités sont intuitives et bien documentées.

2. **Personnalisable :** Vous pouvez personnaliser presque tous les aspects de vos graphiques, de l'axe des x et y aux légendes et couleurs.

3. **Compatibilité avec NumPy et Pandas :** Matplotlib s'intègre bien avec les autres bibliothèques populaires de Python pour l'analyse de données, comme NumPy et Pandas.

4. **Large Gamme de Graphiques :** Il prend en charge différents types de graphiques tels que les graphiques en barres, les histogrammes, les graphiques à dispersion, les graphiques en boîte, etc.

5. **Graphiques de Qualité de Publication :** Matplotlib est souvent utilisé pour créer des graphiques de qualité de publication qui sont utilisés dans des articles scientifiques et des rapports.

**Les Principaux Composants de Matplotlib :**

1. **Figure :** La figure est le canevas global sur lequel vous pouvez placer plusieurs axes et éléments.

2. **Axes :** Les axes représentent les zones où les données sont effectivement tracées. Chaque figure peut avoir plusieurs axes.

3. **Lignes/Points/Barres :** Ce sont les éléments de base que vous utilisez pour représenter vos données.

4. **Légendes :** Elles fournissent des informations supplémentaires sur les éléments du graphique.

<h4>Nuage de point</h4>

 <h4>Nuage de Points</h4>
    <p>Un nuage de points, également connu sous le nom de diagramme de dispersion, est un type de graphique utilisé en statistiques et en visualisation de données. Il représente la distribution de deux variables numériques et permet de visualiser la relation entre elles.</p>
    <h5>Éléments Clés :</h5>
    <ul>
        <li><strong>Axes :</strong> Un nuage de points est créé sur un système d'axes cartésiens avec deux axes : l'axe horizontal (X) et l'axe vertical (Y).</li>
        <li><strong>Points :</strong> Chaque point sur le graphique représente une paire de valeurs correspondant aux deux variables.</li>
        <li><strong>Distribution :</strong> Le nuage de points donne une idée de la distribution des données et de la tendance générale.</li>
        <li><strong>Corrélation :</strong> Il permet d'observer s'il existe une corrélation entre les deux variables.</li>
        <li><strong>Tendances :</strong> Si les points forment une ligne ou une tendance générale, cela indique une relation linéaire entre les variables.</li>
        <li><strong>Dispersion :</strong> La dispersion des points autour de la tendance indique la variabilité des données.</li>
        <li><strong>Outliers :</strong> Les points qui se trouvent loin du groupe principal peuvent être des valeurs aberrantes (outliers) et méritent une attention particulière.</li>
    </ul>

```python
fig = plt.figure(figsize= (20,10))
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.scatter(x, y, color='black')
```

<h4>Histogramme</h4>


<h4>Histogramme</h4>
    <p>L'histogramme est un type de graphique utilisé en statistiques pour représenter la distribution d'une variable quantitative. Il est particulièrement utile pour visualiser la fréquence des différentes valeurs ou plages de valeurs.</p>
    <h5>Éléments Clés :</h5>
    <ul>
        <li><strong>Plages (ou classes) :</strong> Les valeurs sont regroupées en plages ou classes. Chaque barre de l'histogramme représente une plage de valeurs.</li>
        <li><strong>Fréquences :</strong> La hauteur de chaque barre représente le nombre d'observations qui tombent dans la plage correspondante.</li>
        <li><strong>Axes :</strong> L'axe horizontal (X) représente les plages de valeurs, tandis que l'axe vertical (Y) représente les fréquences.</li>
        <li><strong>Largeur des barres :</strong> La largeur des barres peut être ajustée pour montrer plus de détails ou une vue plus globale de la distribution.</li>
        <li><strong>Normalisation :</strong> Il est possible de normaliser l'histogramme pour obtenir une distribution de probabilité.</li>
        <li><strong>Aire :</strong> L'aire totale de l'histogramme est proportionnelle au nombre total d'observations.</li>
    </ul>


```python
fig = plt.figure(figsize= (20,10))
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')

# Ajouter des labels et un titre
plt.xlabel('Valeurs')
plt.ylabel('Fréquence')
plt.title('Exemple d\'histogramme')
```

<h4>Diagramme en barre</h4>

Un diagramme en barres, est un type de graphique utilisé pour représenter des données catégoriques avec des barres rectangulaires. Chaque barre représente une catégorie spécifique, et la hauteur de la barre est proportionnelle à la valeur associée à cette catégorie.
Les barplots sont utiles pour comparer visuellement les valeurs entre différentes catégories. Ils sont particulièrement efficaces lorsque vous avez des données discrètes ou catégoriques.

```python
fig = plt.figure(figsize= (20,10))
# Données
categories = ['A', 'B', 'C', 'D']
valeurs = [10, 20, 15, 25]
color = ['red', 'blue', 'green', 'purple']
# Créer le barplot
plt.bar(categories, valeurs, color= color)

# Ajouter des labels et un titre
plt.xlabel('Catégories')
plt.ylabel('Valeurs')
plt.title('Exemple de Barplot')
```

<h4>Box plot</h4>

Le diagramme en boîte, également appelé "boxplot" en anglais, est un type de graphique très utile pour visualiser la distribution des données, ainsi que pour détecter la présence de valeurs aberrantes (outliers). Voici une description de ses éléments clés :

<ul>
    <li>
        <em>Boîte (Box) :</em>
        <p>
            La boîte centrale représente le deuxième (50e percentile) et le troisième quartile (75e percentile) des données. La hauteur de la boîte est appelée l'étendue interquartile (IQR). Plus la boîte est grande, plus la dispersion des données est grande.
        </p>
    </li>
    <li>
        <em>Ligne médiane (Median) :</em>
        <p>
            La ligne à l'intérieur de la boîte représente la médiane des données (50e percentile), c'est-à-dire la valeur qui divise l'ensemble des données en deux parties égales.
        </p>
    </li>
    <li>
        <em>Moustaches (Whiskers) :</em>
        <p>
            Les moustaches s'étendent jusqu'aux valeurs situées à l'extérieur des quartiles. En général, les moustaches représentent 1,5 fois l'IQR à partir des bords de la boîte. Les points situés en dehors de ces moustaches sont souvent considérés comme des valeurs aberrantes.
        </p>
    </li>
    <li>
        <em>Points aberrants (Outliers) : </em>
        <p>
            Ce sont les points qui se trouvent en dehors des moustaches. Ils sont représentés individuellement.
        </p>
    </li>
</ul>

```python
fig = plt.figure(figsize= (20,10))
# Générer des données aléatoires
np.random.seed(0)
donnees = np.random.normal(0, 1, 100)
donnees[99] = 3 
# Créer le diagramme en boîte
plt.boxplot(donnees)

# Ajouter des titres et étiquettes
plt.xlabel('Données')
plt.ylabel('Valeurs')
plt.title('Diagramme en boîte')
```

```python
fig = plt.figure(figsize= (20,10))
# Générer des données aléatoires pour chaque boîte
np.random.seed(0)
donnees1 = np.random.normal(0, 1, 100)
donnees2 = np.random.normal(2, 1, 100)
donnees3 = np.random.normal(4, 1, 100)

# Regrouper les données dans une liste
donnees = [donnees1, donnees2, donnees3]

# Créer le boxplot avec une seule boîte
plt.boxplot(donnees)

# Définir des étiquettes pour chaque boîte
plt.xticks([1, 2, 3], ['Boîte 1', 'Boîte 2', 'Boîte 3'])

# Ajouter un titre
plt.title('Boxplots')
```

<h4>Le pie chart</h4>

Le diagramme circulaire (ou camembert en français) est un type de graphique qui représente des données sous forme d'un cercle divisé en tranches. Chaque tranche représente une catégorie et la taille de chaque tranche est proportionnelle à la quantité ou à la fréquence qu'elle représente.
Les diagrammes circulaires sont souvent utilisés pour représenter des données catégorielles qui se totalisent à 100%. Ils sont utiles pour visualiser la répartition relative des différentes catégories dans un ensemble de données.

```python
fig = plt.figure(figsize= (20,10))

# Données à représenter
categories = ['A', 'B', 'C', 'D']
valeurs = [25, 32, 12, 31]

# Créer le diagramme circulaire
plt.pie(valeurs, labels=categories, autopct='%1.1f%%', startangle=140)

# Ajouter un titre
plt.title('Répartition des catégories')
```

<h4>Multiple figure</h4>

Une figure multiple, en matière de visualisation de données, fait référence à un graphique qui contient plusieurs sous-graphiques ou sous-diagrammes. Ces sous-graphiques sont souvent utilisés pour représenter différentes perspectives ou détails d'un ensemble de données dans un même espace visuel.

Dans une figure multiple, chaque sous-graphique (ou axe) est placé dans une grille, et chaque axe peut afficher une visualisation différente. Cela permet de comparer facilement les données ou les tendances entre les sous-graphiques.

Par exemple, vous pourriez avoir une figure multiple avec une ligne de temps en haut, un histogramme en bas à gauche, et un nuage de points en bas à droite, chacun montrant des aspects différents des mêmes données.

L'utilisation de figures multiples est courante lorsqu'on veut explorer et comparer plusieurs aspects des données simultanément, ce qui peut être très utile pour obtenir une compréhension globale d'un ensemble de données complexe.

```python
import matplotlib.pyplot as plt

# Créer une figure et des sous-graphiques
fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (15, 10))

# Remplir les sous-graphiques avec des données
axes[0, 0].plot([1, 2, 3, 4], [10, 5, 20, 15])
axes[0, 1].scatter([1, 2, 3, 4], [10, 5, 20, 15])
axes[1, 0].bar(['A', 'B', 'C'], [3, 6, 8])
axes[1, 1].hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])

# Ajouter des titres aux sous-graphiques
axes[0, 0].set_title('Graphique 1')
axes[0, 1].set_title('Graphique 2')
axes[1, 0].set_title('Graphique 3')
axes[1, 1].set_title('Graphique 4')

# Ajuster la disposition des sous-graphiques
plt.tight_layout()

# Afficher la figure
plt.show()
```

<h3>Style</h3>

En analyse de données, le terme "style" fait référence à la manière dont les données sont présentées visuellement. Il englobe l'ensemble des choix esthétiques et graphiques qui permettent de représenter les données de manière claire et compréhensible. Cela inclut des éléments tels que les couleurs, les types de graphiques, les étiquettes, les légendes, les titres, les axes, etc.

<em>Avantages des styles</em>

<ul>
    <li><strong>Clarté et Lisibilité :</strong> Un style bien pensé rend les graphiques et visualisations plus clairs et faciles à lire, ce qui facilite la compréhension des informations présentées.</li>
    <li><strong>Mise en Évidence des Tendances :</strong> Un bon style peut mettre en évidence les tendances et les patterns dans les données, permettant ainsi aux observateurs de saisir rapidement les insights les plus importants.</li>
    <li><strong>Amélioration de la Compréhension :</strong> Un style approprié peut rendre les données plus compréhensibles pour un public plus large, même pour ceux qui ne sont pas familiers avec les détails techniques.</li>
    <li><strong>Personnalisation :</strong> Le style permet de personnaliser les graphiques en fonction des préférences de l'audience et des besoins de l'analyse, ce qui facilite la communication des résultats.</li>
    <li><strong>Impact Visuel :</strong> Un style visuellement attrayant peut susciter l'intérêt de l'audience et encourager une exploration plus approfondie des données.</li>
    <li><strong>Réduction de la Confusion :</strong> Un style bien choisi peut réduire la possibilité de confusion ou de mauvaise interprétation des données, en mettant en évidence les informations les plus pertinentes.</li>
    <li><strong>Adaptabilité :</strong> Le style peut être adapté en fonction du contexte de l'analyse et du public cible, permettant ainsi de mieux répondre aux besoins spécifiques de chaque situation.</li>
    <li><strong>Professionnalisme :</strong> Un style soigné et bien exécuté renforce la crédibilité de l'analyse et donne une impression professionnelle, ce qui est crucial dans un environnement d'affaires.</li>
    <li><strong>Facilitation de la Prise de Décision :</strong> En mettant en évidence les principaux résultats, un style approprié peut aider à prendre des décisions éclairées basées sur les données.</li>
</ul>


<em>Quelques fonctions pour définir le style</em>

<ol>
<li><code>plt.style.use(style_name)</code>: Cette fonction permet de sélectionner un style prédéfini pour le graphique. Par exemple : 'seaborn', 'ggplot', 'classic', etc.</li>
<li><code>plt.title(label, fontsize, fontweight, color)</code>: Permet de définir le titre du graphique avec des options de personnalisation telles que la taille de la police, le poids de la police et la couleur.</li>
<li><code>plt.xlabel(label, fontsize, fontweight, color)</code>: Définit le label de l'axe des x avec des options de personnalisation.</li>
<li><code>plt.ylabel(label, fontsize, fontweight, color)</code>: Définit le label de l'axe des y avec des options de personnalisation.</li>
<li><code>plt.xticks(ticks, labels, fontsize, fontweight, color)</code>: Personnalise les marques (ticks) sur l'axe des x avec des options de personnalisation.</li>
<li><code>plt.yticks(ticks, labels, fontsize, fontweight, color)</code>: Personnalise les marques (ticks) sur l'axe des y avec des options de personnalisation.</li>
<li><code>plt.grid(bool)</code>: Active ou désactive la grille du graphique.</li>
<li><code>plt.legend()</code>: Ajoute une légende au graphique.</li>
<li><code>plt.xlim(min, max)</code>: Définit les limites de l'axe des x.</li>
<li><code>plt.ylim(min, max)</code>: Définit les limites de l'axe des y.</li>
<li><code>plt.figure(figsize=(width, height))</code>: Définit la taille de la figure.</li>
<li><code>plt.subplots(rows, cols)</code>: Permet de créer des sous-graphiques (subplots).</li>
</ol>
